```python
import pandas as pd
import numpy as np
```

```python
noti_path = 'csv/search_noti_02.csv'
search_path = 'csv/item_search_02.csv'
# click_path = 'csv/item_click_0301.csv'

noti_df = pd.read_csv(noti_path).dropna(axis=0)
search_df = pd.read_csv(search_path).dropna(axis=0)
# click_df = pd.read_csv(click_path).dropna(axis=0)
```

```python
# print(noti_df.dtypes)
# print(search_df.dtypes)

noti_df = noti_df[noti_df.uid > 0]
search_df = search_df[search_df.viewer_uid > 0]
```

```python
noti_search_df = pd.merge(noti_df, search_df, how='left', left_on=['uid', 'keyword'], right_on=['viewer_uid', 'keyword'])

noti_search_df.to_csv('csv/noti_search_df.csv', mode='w', header=True)

noti_search_df.tail(10)
```

```python

# noti_search_df['latest_notified'] = pd.to_datetime(noti_search_df['latest_notified'], errors='coerce')
# noti_search_df['created_at'] = pd.to_datetime(noti_search_df['created_at'], errors='coerce')
# noti_search_df['updated'] = pd.to_datetime(noti_search_df['updated'], errors='coerce')

def between(row):
    
    latest = row['latest_notified']
    created = row['created_at']
    updated = row['updated']
    
    latest = pd.to_datetime(latest, errors = 'coerce')
    created = pd.to_datetime(created, errors = 'coerce')
    updated = pd.to_datetime(updated, errors = 'coerce')
    
    if updated >= created and updated <= latest:
        between = 1
    else:
        between = 0
        
    return between
    
noti_search_df.loc[:,'between'] = noti_search_df[['latest_notified', 'created_at', 'updated']].apply(between, axis = 1)

noti_search_df.tail(10)
```

```python
noti_search_df.to_csv('csv/noti_search_df_1.csv', index=False, mode='w', header=True)
```
