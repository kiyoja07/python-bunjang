```python
import pandas as pd
```

```python
# 등록된 제품 수 by category, date
read_path = '../csv/products.csv'
df_product = pd.read_csv(read_path)

df_product.tail()

# status : 0: 판매중, 1: 예약됨, 2: 삭제, 3: 판매완료
```

```python
df_product_pivot = df_product.pivot_table(index=['category_id', 'on_at', 'close_at'],
                                          columns='status', 
                                          values='products',
                                          dropna=False)

df_product_pivot = df_product_pivot.fillna(0)

df_product_pivot.tail()
```

```python
df_product_pivot['on_sale'] = df_product_pivot.sum(axis=1)

df_product_pivot['close_sale'] = df_product_pivot['on_sale'] - df_product_pivot[0]

df_product_pivot.tail()
```
