```python
status_delete = status[status['status'] == 2]

status_delete
```

```python
status_delete.groupby(['pid']).agg({'updated': 'count'})
```

```python
status_complete = status[status['status'] == 3]

status_complete
```

```python
bunp_delete = pd.merge(bunp, status_delete, on='pid', how='left')

bunp_delete_complete = pd.merge(bunp_delete, status_complete, on='pid', how='left')

bunp_delete_complete = bunp_delete_complete.rename(columns={'updated_x': 'updated_delete', 'updated_y': 'updated_complete'})

bunp_delete_complete
```

```python
bunp_delete_complete_result = bunp_delete_complete.groupby(['pid'], as_index=False).agg({'updated_delete': pd.Series.nunique, 'updated_complete': pd.Series.nunique})

bunp_delete_complete_result
```
