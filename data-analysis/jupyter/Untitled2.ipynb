```python
import pandas as pd
```

```python
## import search term file
read_path = '../csv/search_term_200501_200510.csv'
search_term = pd.read_csv(read_path)
search_term = search_term.rename(columns={'count': 'search_count'})
search_term['search_rank'] = search_term['search_count'].rank(ascending=False, method = 'max')
search_term.head()
```

```python
## import search click file
read_path = '../csv/search_click_200501_200510.csv'
search_click = pd.read_csv(read_path)
search_click.head()
```

```python
term_click = pd.merge(search_term, search_click, on='ref_term',how='left')
term_click_1000 = term_click[term_click['search_rank'] <= 1000]

term_click_1000['category'] = term_click_1000['category'].astype(str)
term_click_1000['category_2th'] = term_click_1000['category'].str[:6]
term_click_1000['category_1th'] = term_click_1000['category'].str[:3]
term_click_1000
```

```python
term_click_1000_result = term_click_1000.groupby(['search_rank', 
                                                  'ref_term', 
                                                  'category_1th'], as_index=False).agg({'search_count': 'mean', 'click': 'sum'})
term_click_1000_result
```

```python
# save
save_path = '../csv/result/search_term_click_200501_200510.csv'
term_click_1000_result.to_csv(save_path, index=False, mode='w', header=True)
```
