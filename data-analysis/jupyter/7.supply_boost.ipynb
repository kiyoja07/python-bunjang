```python
import pandas as pd
```

### ì¹´í…Œê³ ë¦¬ íŒŒì¼ ì½ê¸°

```python
# ì¹´í…Œê³ ë¦¬

path = '../csv/categories.csv'

df_category = pd.read_csv(path)

df_category['category'] = df_category['category'].astype(str)

df_category = df_category[['category', 'name']]

df_category.tail()
```

### í´ë¦­ íŒŒì¼ ì½ê¸°

```python
# ì„±, ì—°ë ¹, ì¹´í…Œê³ ë¦¬ì— ë”°ë¥¸ ì›”ë³„ í´ë¦­ ìˆ˜
read_path = '../csv/clicked_cohort.csv'
df_click = pd.read_csv(read_path)

df_click.tail()
```

```python
df_click['category_6'] = df_click['category_id'].astype(str).str[:6] # 2 depth category

df_click_6 = df_click.groupby(['month_at', 'category_6', 'age', 'gender'], as_index=False).agg({'clicked': 'sum'})

df_click_6.tail()
```

```python
# ì¹´í…Œê³ ë¦¬ ë¶™ì´ê¸°
df_click_cate_6 = pd.merge(df_click_6, df_category[['category', 'name']], 
                           how='inner', left_on=['category_6'], right_on=['category'])

df_click_cate_6 = df_click_cate_6.drop(['category_6'], axis=1).rename(columns={'name': 'category_name'})

df_click_cate_6.tail()
```

```python
# save
save_path = '../csv/tmp/df_click_cate_6.csv'
df_click_cate_6.to_csv(save_path, index=False, mode='w', header=True)
```

```python
read_path = '../csv/products_cohort.csv'
df_products = pd.read_csv(read_path)

df_products = df_products.drop(['status'], axis=1)

df_products.tail()
```

### í´ë¦­ê³¼ ë²ˆí”„ íŒŒì¼ í•©ì¹˜ê¸°

```python
# í´ë¦­ íŒŒì¼

read_path = '../csv/tmp/df_click_cate_6.csv'
df_click_cate_6 = pd.read_csv(read_path)

df_click_cate_6.tail()
```

```python
# ë²ˆí”„ íŒŒì¼ ì¤‘ êµ¬ë§¤ì ë°ì´í„°

df_bunp_buyer = df_bunp.groupby(['category_id', 'month_at', 'buyer_age', 'buyer_gender'], as_index=False).agg({'selled_count': 'sum', 'bunp_amount': 'sum'})

df_bunp_buyer.tail()
```
